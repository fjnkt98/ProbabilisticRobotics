```python
import sys
sys.path.append("../scripts/")
from mcl import *
```

```python
def trial():
    time_interval = 0.1
    world = World(30, time_interval, False)
    
    m = Map()
    for ln in [(-4, 2), (2, -3), (3, 3)]:
        m.append_landmark(Landmark(*ln))
    world.append(m)
    
    initial_pose = np.array([np.random.uniform(-5.0, 5.0), np.random.uniform(5.0, -5.0), np.random.uniform(-math.pi, math.pi)]).T
    robot_pose = np.array([np.random.uniform(-5.0, 5.0), np.random.uniform(5.0, -5.0), np.random.uniform(-math.pi, math.pi)]).T
    estimator = Mcl(m, initial_pose, 100)
    agent = EstimationAgent(time_interval, 0.2, 10.0/180*math.pi, estimator)
    r = Robot(robot_pose, sensor=Camera(m), agent=agent, color="red")
    world.append(r)
    
    world.draw()
```

```python
trial()
```
